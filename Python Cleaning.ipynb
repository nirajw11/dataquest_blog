```python
import re

def clean_split_dates(row):
    initial_date=str(row['Date']) 
    
    split_date = re.split('-|–',initial_date)
    
    if len(split_date) > 1:
        proc_date = split_date[0]
    else:
        proc_date = initial_date
    
    split_date=proc_date.split(' ')
    if len(split_date) > 1:
        new_date = split_date[1]
        new_date=new_date[0:4]
    else:
        new_date = proc_date
    
    split_date=new_date.split('.')
    if len(split_date) > 1:
        final_date = split_date[1]
        final_date=final_date[0:4]
    else:
        final_date = new_date
    
    if len(final_date) < 4:
        final_date=0
    elif len(final_date) == 5:
        final_date=final_date[0:4]
    elif len(final_date) !=4:
        final_date='Unknown'
    
    return final_date
```

```python
artworks['Date'] = artworks.apply(lambda row : clean_split_dates(row),axis=1)
```
